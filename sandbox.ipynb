# Testing Null Code

```python
import pandas as pd
import numpy as np
from data_frame_info import DataFrameInfo as dfi

loandata_df = pd.read_csv("loan_data2.csv")

null_percentage = (loandata_df.isnull().sum() / len(loandata_df)) * 100

print(null_percentage)
```

```python
threshold = 50
# null_percent = dfi.df_null_count(loandata_df)
column_drop = null_percentage[null_percentage > threshold].index
print(column_drop)
loandata_df_drop = loandata_df.drop(columns = column_drop)
loandata_df_drop.head(10)
```

```python
loandata_df_drop.skew(axis = 0, skipna = True, numeric_only = True)
```

```python
import matplotlib.pyplot as plt

numeric_columns = loandata_df_drop.select_dtypes(include=['number']).columns

df_numeric = loandata_df_drop[numeric_columns]
corr = df_numeric.corr()
print(corr)
# plt.matshow(corr)
# plt.show()
corr.style.background_gradient(cmap='coolwarm')
```
